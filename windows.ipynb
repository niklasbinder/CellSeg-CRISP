```python
%matplotlib qt
from multiprocessing import Process
from main_plane import main

def run_single(file_index, IF, GM, GQ, BT):
  main(file_index, IF, GM, GQ, BT)

def run():
  for region in [1,2,3,4]:
    for IF in [3]:
      for GM in [0,-1]:
        for GQ in [2,3]:
          for BT in [1,1.5,2,2.5]:
            p = Process(target=run_single, args=(region-1, IF, GM, GQ, BT,))
            p.start()
            p.join()

if __name__ == '__main__':
  #run()
  main(0, 3.0, -1, 2, 1.5)
```
